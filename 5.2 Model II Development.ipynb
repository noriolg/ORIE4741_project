```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.metrics import explained_variance_score, mean_squared_error, r2_score
# Import the model we are using
from sklearn.datasets import make_regression
from sklearn.ensemble import RandomForestRegressor
```

# Load Dataset for 5 large neighborhoods

```python
# 2020
df_train_x_2020_L = pd.read_excel("data/training_testing_5_large_neighborhood_sophie_version/X_train20_L.xlsx");
df_train_y_2020_L = pd.read_excel("data/training_testing_5_large_neighborhood_sophie_version/y_train20_L.xlsx");
df_test_x_2020_L = pd.read_excel("data/training_testing_5_large_neighborhood_sophie_version/X_test20_L.xlsx");
df_test_y_2020_L = pd.read_excel("data/training_testing_5_large_neighborhood_sophie_version/y_test20_L.xlsx");

# 2019
df_train_x_2019_L = pd.read_excel("data/training_testing_5_large_neighborhood_sophie_version/X_train19_L.xlsx");
df_train_y_2019_L = pd.read_excel("data/training_testing_5_large_neighborhood_sophie_version/y_train19_L.xlsx");
df_test_x_2019_L = pd.read_excel("data/training_testing_5_large_neighborhood_sophie_version/X_test19_L.xlsx");
df_test_y_2019_L = pd.read_excel("data/training_testing_5_large_neighborhood_sophie_version/y_test19_L.xlsx");
```

Need to drop id column, otherwise random forest will not work on overflow or NaN values

```python
# 2020
df_train_x_2020_L = df_train_x_2020_L.drop(['id'], axis=1)
df_test_x_2020_L = df_test_x_2020_L.drop(['id'], axis=1)
df_train_y_2020_L = df_train_y_2020_L['price']
df_test_y_2020_L = df_test_y_2020_L['price']
# 2019
df_train_x_2019_L = df_train_x_2019_L.drop(['id'], axis=1)
df_test_x_2019_L = df_test_x_2019_L.drop(['id'], axis=1)
df_train_y_2019_L = df_train_y_2019_L['price']
df_test_y_2019_L = df_test_y_2019_L['price']
```

# Conduct Random forest Regression for 5 large neighborhoods data


```python
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
```

```python
train_x_2020_L = np.array(df_train_x_2020_L)
train_y_2020_L = np.array(df_train_y_2020_L)
test_x_2020_L = np.array(df_test_x_2020_L)
test_y_2020_L = np.array(df_test_y_2020_L)

train_x_2019_L = np.array(df_train_x_2019_L)
train_y_2019_L = np.array(df_train_y_2019_L)
test_x_2019_L = np.array(df_test_x_2019_L)
test_y_2019_L = np.array(df_test_y_2019_L)
```

```python
rf.fit(train_x_2020_L, train_y_2020_L);
```

```python
rand_forest_start = time.time()
# Use the forest's predict method on the test data
training_preds_20_L = rf.predict(train_x_2020_L)
val_preds_20_L = rf.predict(test_x_2020_L)
rand_forest_end = time.time()
print(str((rand_forest_end - rand_forest_start)/60) + " minutes")
print("\nTraining MSE:", round(mean_squared_error(train_y_2020_L, training_preds_20_L),4))
print("Validation MSE:", round(mean_squared_error(test_y_2020_L, val_preds_20_L),4))
print("\nTraining r2:", round(r2_score(train_y_2020_L, training_preds_20_L),4))
print("Validation r2:", round(r2_score(test_y_2020_L, val_preds_20_L),4))
```

```python
rf.fit(train_x_2019_L, train_y_2019_L);
```

```python
rand_forest_start = time.time()
# Use the forest's predict method on the test data
training_preds_19_L = rf.predict(train_x_2019_L)
val_preds_19_L = rf.predict(test_x_2019_L)
rand_forest_end = time.time()
print(str((rand_forest_end - rand_forest_start)/60) + " minutes")
print("\nTraining MSE:", round(mean_squared_error(train_y_2019_L, training_preds_19_L),4))
print("Validation MSE:", round(mean_squared_error(test_y_2019_L, val_preds_19_L),4))
print("\nTraining r2:", round(r2_score(train_y_2019_L, training_preds_19_L),4))
print("Validation r2:", round(r2_score(test_y_2019_L, val_preds_19_L),4))
```

# Load Dataset for 200 small neighborhoods

```python
# 2020
df_train_x_2020 = pd.read_excel("data/training_testing_200_neighborhood_sophie_version/X_train20.xlsx");
df_train_y_2020 = pd.read_excel("data/training_testing_200_neighborhood_sophie_version/y_train20.xlsx");
df_test_x_2020 = pd.read_excel("data/training_testing_200_neighborhood_sophie_version/X_test20.xlsx");
df_test_y_2020 = pd.read_excel("data/training_testing_200_neighborhood_sophie_version/y_test20.xlsx");

# 2019
df_train_x_2019 = pd.read_excel("data/training_testing_200_neighborhood_sophie_version/X_train19.xlsx");
df_train_y_2019 = pd.read_excel("data/training_testing_200_neighborhood_sophie_version/y_train19.xlsx");
df_test_x_2019 = pd.read_excel("data/training_testing_200_neighborhood_sophie_version/X_test19.xlsx");
df_test_y_2019 = pd.read_excel("data/training_testing_200_neighborhood_sophie_version/y_test19.xlsx");
```

Need to drop id column, otherwise random forest will not work on overflow or NaN values

```python
# 2020
df_train_x_2020 = df_train_x_2020.drop(['id'], axis=1)
df_test_x_2020 = df_test_x_2020.drop(['id'], axis=1)
df_train_y_2020 = df_train_y_2020['price']
df_test_y_2020 = df_test_y_2020['price']
# 2019
df_train_x_2019 = df_train_x_2019.drop(['id'], axis=1)
df_test_x_2019 = df_test_x_2019.drop(['id'], axis=1)
df_train_y_2019 = df_train_y_2019['price']
df_test_y_2019 = df_test_y_2019['price']
```

```python
train_x_2020 = np.array(df_train_x_2020)
print("1")
train_y_2020 = np.array(df_train_y_2020)
print("2")
test_x_2020 = np.array(df_test_x_2020)
print("3")
test_y_2020 = np.array(df_test_y_2020)
print("4")


train_x_2019 = np.array(df_train_x_2019)
print("5")
train_y_2019 = np.array(df_train_y_2019)
print("6")
test_x_2019 = np.array(df_test_x_2019)
print("7")
test_y_2019 = np.array(df_test_y_2019)
print("8")
```

# Conduct Random forest Regression for 200 large neighborhoods data


2020 random forest

```python
random_states = [5, 10, 25, 50, 100, 150, 200, 225]
num_estimators = [500, 1000, 1500, 2000, 2500, 3000]
for i in range(len(random_states)):
    for j in range(len(num_estimators)):
        rand_forest_start = time.time()
        random_state = random_states[i]
        n_estimator =  num_estimators[j]
        rf = RandomForestRegressor(n_estimators = n_estimator, random_state = random_state)
        rf.fit(train_x_2020, train_y_2020);
        # Use the forest's predict method on the test data
        training_preds_20 = rf.predict(train_x_2020)
        val_preds_20 = rf.predict(test_x_2020)
        rand_forest_end = time.time()
        print("************Training with " + str(random_state) + " random states and "+ str(n_estimator) + " decision trees ************")
        print(str((rand_forest_end - rand_forest_start)/60) + " minutes")
        print("\nTraining MSE:", round(mean_squared_error(train_y_2020, training_preds_20),4))
        print("Validation MSE:", round(mean_squared_error(test_y_2020, val_preds_20),4))
        print("\nTraining r2:", round(r2_score(train_y_2020, training_preds_20),4))
        print("Validation r2:", round(r2_score(test_y_2020, val_preds_20),4))
```

Check important features for 3000 trees and 10 random features in each split

```python
rf_2020 = RandomForestRegressor(n_estimators = 3000, random_state = 10)
rf_2020.fit(train_x_2020, train_y_2020);
ft_weights_randfor20 = pd.DataFrame(rf_2020.feature_importances_, columns=['weight'], index=df_train_x_2020.columns)
ft_weights_randfor20.sort_values('weight', inplace=True)

plt.figure(figsize=(8,50))
plt.barh(ft_weights_randfor20.index, ft_weights_randfor20.weight, align='center') 
plt.title("Feature importances in the Random forest model (3000 trees, 10 features)", fontsize=14)
plt.xlabel("Feature importance")
plt.margins(y=0.01)
plt.show()
```

```python
rf_2020 = RandomForestRegressor(n_estimators = 3000, random_state = 150)
rf_2020.fit(train_x_2020, train_y_2020);
ft_weights_randfor20 = pd.DataFrame(rf_2020.feature_importances_, columns=['weight'], index=df_train_x_2020.columns)
ft_weights_randfor20.sort_values('weight', inplace=True)

plt.figure(figsize=(8,50))
plt.barh(ft_weights_randfor20.index, ft_weights_randfor20.weight, align='center') 
plt.title("Feature importances in the Random forest model (3000 trees, 150 features)", fontsize=14)
plt.xlabel("Feature importance")
plt.margins(y=0.01)
plt.show()
```
