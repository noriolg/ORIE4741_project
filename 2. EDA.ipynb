```julia
# bring packages into main namespace
using DataFrames             # Data tables are called "DataFrames"
using StatsPlots             # load plotting packages 
using Statistics             # basic statistical functions
using CSV                    # tools for working with CSV files
# plotly() # good interactive html plotting tool
pyplot()   # good plotting tool for saving pdf figures
```

# Histograms for April - September data for final report

```julia
df = CSV.read("./data/airbnb_April_Sep2019.csv");
unique(df[:room_type])
```

```julia
@df df histogram(:price, group=:room_type, label=["entire home" "private room" "shared room" "hotel room"])
plot!(xlims=(0,1500))
xlabel!("Price Distribution based on Room type before Covid-19")
```

price distribution after covid-19

```julia
df = CSV.read("./data/airbnb_April_Sep2020.csv");
unique(df[:room_type])
```

```julia
@df df histogram(:price, group=:room_type, label=["entire home" "private room" "shared room" "hotel room"])
plot!(xlims=(0,1500))
xlabel!("Price Distribution based on Room type after Covid-19")
```

# Histograms for midterm report 

```julia
df = CSV.read("./data/treated data/2019_july_listings_encoded.csv");
df
unique(df[:room_type])
unique(df[:neighbourhood_group])
```

```julia
histogram(df[:price], label="price")
plot!(xlims=(0,1000))
xlabel!("Price")
```

```julia
@df df histogram(:price, group=:room_type, label=["private room" "entire home" "shared room"])
plot!(xlims=(0,1000))
xlabel!("Prices in June 2019")
```

```julia
@df df histogram(:price, group=:neighbourhood_group, label=["Brooklyn" "Manhattan" "Queens" "Staten Island" "Bronx"])
plot!(xlims=(0,1000))
xlabel!("Prices in June 2019")
```
